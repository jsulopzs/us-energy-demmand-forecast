# Prepare data for modelling each region

The data will be spilt into:

1. `train`: 85% of the data to train and validate the model
2. `prod`: 15% of the data to simulate the production environment

## Data

Initially, we load all the regions to later create data files for each region.

```python
import pandas as pd

df_regions = pd.read_csv('../data/general/pjm_pivot.csv', index_col=0, parse_dates=True)
df_regions.head()
```

## Train Prod Split

```python
from sklearn.model_selection import train_test_split
import os

for region in df_regions.columns:
    
    df = df_regions[region].to_frame()
    train, prod = train_test_split(df, test_size=0.15, shuffle=False)

    path_region = f'../data/regions/{region}'
    if not os.path.exists(path_region):
        os.makedirs(path_region)

    for df, name in zip([train, prod], ['train', 'prod']):
        df.to_csv(f'{path_region}/{name}.csv')
```
